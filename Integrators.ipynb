```python
temp, time, kinetic_energy = np.loadtxt('cmake-build-debug/vel_verlet_etot.dat', unpack=True)
```

```python
plt.plot(temp,kinetic_energy,label="Velocity-Verlet")
plt.xlabel('T')
plt.ylabel('K.E')
plt.legend(loc=2)
plt.savefig("Temp_Energy.pdf")
plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def func(x,a):
    return a * ( x)
```

```python
temp = 1/beta
```

```python
popt, pcov = curve_fit(func, beta, kinetic_energy)
```

```python
plt.plot(temp, func(beta, *popt), 'r-', label='fit')
plt.plot(temp,kinetic_energy,label="Velocity-Verlet")
plt.xlabel('T')
plt.ylabel('K.E')
plt.legend(loc=2)
plt.savefig("Temp_Energy.pdf")
plt.show()
```
